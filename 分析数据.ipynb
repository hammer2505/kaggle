```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from pylab import *

import warnings
warnings.filterwarnings('ignore')
```

# read data

```python
filename = "data/train.csv"
data = pd.read_csv(filename)
data.set_index('id', inplace=True)

for a in range(1,117):
    a = "cat" + str(a)
    data[a] = data[a].astype('category')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
```

```python
contFeatureslist = []
for a in range(1,15):
    a = "cont" + str(a)
    contFeatureslist.append(a)
```

```python
plt.figure(figsize=(13,9))
sns.boxplot(data[contFeatureslist])
```

# Correlation between continuous features

```python
contFeatureslist.append("loss")
```

```python
correlationMatrix = data[contFeatureslist].corr().abs()

plt.subplots(figsize=(13, 9))
sns.heatmap(correlationMatrix,annot=True)

# Mask unimportant features
sns.heatmap(correlationMatrix, mask=correlationMatrix < 1, cbar=False)
plt.show()
```

# Analysis of loss feature

```python
plt.figure(figsize=(13,9))
sns.distplot(data["loss"])
sns.boxplot(data["loss"])
```

```python
plt.figure(figsize=(13,9))
sns.distplot(np.log1p(data["loss"]))
```
