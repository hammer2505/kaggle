```python
# Import all necessary modules.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from pylab import *

import warnings
warnings.filterwarnings('ignore')
```

```python
# read data
filename = "../data/train.csv"
traindata = pd.read_csv(filename)
traindata.set_index('id', inplace=True)

for a in range(1,117):
    a = "cat" + str(a)
    traindata[a] = traindata[a].astype('category')
    
filename = "../data/test.csv"
testdata = pd.read_csv(filename)
testdata.set_index('id', inplace=True)

for a in range(1,117):
    a = "cat" + str(a)
    testdata[a] = testdata[a].astype('category')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
```

```python
contFeatureslist = []
for a in range(1,15):
    a = "cont" + str(a)
    contFeatureslist.append(a)
```

```python
plt.figure(figsize=(13,9))
sns.boxplot(traindata[contFeatureslist])
```

```python
plt.figure(figsize=(13,9))
sns.boxplot(testdata[contFeatureslist])
```

# Correlation between continuous features

```python
correlationMatrix = testdata[contFeatureslist].corr().abs()

plt.subplots(figsize=(13, 9))
sns.heatmap(correlationMatrix,annot=True)

# Mask unimportant features
sns.heatmap(correlationMatrix, mask=correlationMatrix < 1, cbar=False)
plt.show()
```
